## Leitura das bases

```python
# Local do diretório
Local_Diretorio = 'C:/Users/Isabela Vitoriano/Desktop/projeto_analise_prf_multas/infrações_2022/ajustados_2022'
Diretorio_Path = Path(Local_Diretorio)

type(Diretorio_Path)
```

```python
# Lista os arquivos
Arquivos = Diretorio_Path.iterdir()

# Função para calcular o tamanho do  arquivo
def Bytes_Mega(bytes):
    Megabytes = bytes / (1024 * 1024)
    return Megabytes

# Total MBs
Total_MB = 0

# Loop
for Loop in Arquivos:

    # Nome do arquivo
    Nome_Arquivo = str( Loop ).split('\\')[-1]

    # Buscando o tamanho do arquivo
    Tamanho = os.path.getsize(Loop)

    # Convertendo em Megas
    Conversao = round(Bytes_Mega(Tamanho) , 2)

    # Calculando tamanho total da pasta
    Total_MB = round( (Conversao + Total_MB) , 2 )

    # Exibindo  nome e tamanho do arquivo
    print(f'{Nome_Arquivo} : {Conversao} MB')
    print('-'*40)

print(f'Tamanho total dos arquivos: {Total_MB} MB')
```

```python
# Loop para consolidar os arquivos
Arquivos = Diretorio_Path.iterdir()

for Contador, Arq in enumerate(Arquivos):
    print(Contador, Arq)

    # Se for primeira iteração, cria a base consolidada
    if Contador == 0:
        # Ler o arquivo do Loop
        Base_Dados = pd.read_csv(Arq, encoding = 'latin1', sep = ';')
        # Empilhar os dados
        Base_Consolidada = Base_Dados
    else:
        # Ler o arquivo do Loop
        Base_Dados = pd.read_csv(Arq, encoding = 'latin1', sep = ';')
        # Empilhar os dados
        Base_Consolidada = pd.concat( [Base_Consolidada, Base_Dados])
```

```python
# SQL
import sqlite3
```

```python
# Criar o banco de dados
Conexao = sqlite3.connect('Banco_Dados.db')

# Nome da tabela
Nome_Tabela = 'Base_Consolidada'
```

```python
Base_Consolidada.to_sql(Nome_Tabela, Conexao, if_exists='append', index=False)
```

```python
# Parquet
Base_Consolidada.to_parquet('Base_Consolidada.parquet', index=False, compression = 'gzip')
```
